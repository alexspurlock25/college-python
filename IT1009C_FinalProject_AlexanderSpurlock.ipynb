```python
import pandas
from pandas import DataFrame

# these are the files from the dataset from Kaggle
hero_network_file_path = 'marvel/hero-network.csv'
edges_file_path = 'marvel/edges.csv'
nodes_file_path = 'marvel/nodes.csv'
```

```python
# Here is a dictionary for the user selection options
node_options: dict[int, str] = {
    1: 'Search for hero\'s network of heros and villians.',
    2: 'Search for hero\'s comic book appearance.',
    3: 'To exit!'
}
```

```python
def read_csv(file_path: str, cols: list[str]) -> DataFrame: 
    """
    Read a given csv files. This is just so that we don't have to keep typing 'pandas.read_csv' over and over again.
    """
    return pandas.read_csv(file_path, names=cols)
```

```python
def user_type_options_menu():
    """
    Display the menu dictionary in a pretty way.
    """
    print("\n-- MENU --")
    for obj in node_options:
        print(obj, ' - ', node_options[obj])
```

```python
def search_hero_comic_appearance(hero: str):
    """
    Search the dataset file, edges.csv, for a referece for given search req. Then, display the results in a table.
    """
    df = read_csv(edges_file_path, cols=['hero', 'comic book'])
    display(df[df['hero'].str.contains(hero, case=False)])
```

```python
def search_heros_network(hero: str):
    """
    Search the dataset file, hero-network.csv, for a referece for given search req. Then, display the results in a table.
    """
    df = read_csv(hero_network_file_path, cols=['Your Search: ' + hero, hero + '\'s Network']) # modify the col names to make the table prettier
    display(df[df['Your Search: ' + hero].str.contains(hero, case=False)])
```

```python
def main():
    print('Welcome to the Marvel universe comic book/hero/villian search!')

    # loop until the user selects 3
    while True:
        user_type_options_menu() # display the menu every new selectiion
        user_type_choice = int(input('Pick: ')) # pick from 1, 2, or 3

        if user_type_choice == 1:
            user_input = input('Hero name: ') # if user picks 1, search for hero's network of heros and villians
            search_heros_network(user_input)
        if user_type_choice == 2:
            user_input = input('Hero name: ') # if user picks 2, search for hero's comics book appearance
            search_hero_comic_appearance(user_input)
        if user_type_choice == 3:
            print('Goodbye!')
            break # if user picks, exit the loop and say goodbye
    

if __name__ == '__main__':
    main()

# How did this solution solve my problem?
# Well, I am now able to see what heros and villians are bunched up together in the same mini universe, and see what comic book they apear in! Which is great!
# One new thing that I learned here?
# I learned how to rename columns names when it's being displayed in the table. I made my dynamic to whatever the search was.
```
